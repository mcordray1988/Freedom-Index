```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
bottom_five_ethnicity = pd.read_csv('../final-project/data/bottom_5_HIEF_data.csv')
bottom_five_ethnicity
```

```python
congo = bottom_five_ethnicity[(bottom_five_ethnicity.Country == 'Congo')]
congo
```

```python
X = congo.Year.values.reshape(-1, 1)
y = congo.EFindex.values.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(X, y)
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
plt.title("Congo")
plt.xlabel("Years")
plt.ylabel("EF Index")
```
